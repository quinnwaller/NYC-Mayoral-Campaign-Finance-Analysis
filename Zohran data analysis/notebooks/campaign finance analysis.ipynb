```python
import pandas as pd
```

## Zohran Mamdani analysis

To import Zohran Mamdani's campaign finance data from New York City Campaign Finance Board (https://www.nyccfb.info/FTMSearch/Home/FTMSearch) :

```python
mamdani = pd.read_csv("../data/Mamdani data 6.17.csv")
```

```python
mamdani
```

```python
mamdani.columns
```

To filter out donors that live in New York:

```python
filtered_mamdani=mamdani[mamdani["STATE"] !="NY"]
```

To find average amount of individual non-New York donations:

```python
average = filtered_mamdani['AMNT'].mean()

print(average)
```

To find the total amount of non-New York donations:

```python
sum_mamdani = filtered_mamdani["AMNT"].sum()
print(sum_mamdani)
```

To create a new index that only includes the state, amount, and number and amount of non-New York donations by state, including Puerto Rico and Washington D.C.:

```python
mamdani_analysis=filtered_mamdani.groupby("STATE").agg({
    "AMNT": "sum",
    "REFNO": "count"
}).reset_index()
```

```python
mamdani_analysis
```

To rename the column "REFNO" as number of donations:

```python
mamdani_analysis = mamdani_analysis.rename(columns={'REFNO': 'Mamdani number of donations'})
```

```python
mamdani_analysis
```

To output Mamdani Analysis as a CSV file:

```python
mamdani_analysis.to_csv("../outputs/mamdani.6.17.csv")
```
